### Take the natural log of population.

```python
covid = np.log(pd.read_csv("time_series_covid_19_confirmed.csv").drop(["Lat", "Long"], axis=1).groupby("Country/Region").sum() + 1)
covid_de=covid.loc["Germany",:]
covid_it=covid.loc["Italy",:]
covid_it=np.concatenate(([np.linspace(0, len(covid_it)-1, len(covid_it))], [covid_it.values]), axis=0).T
covid_de= np.concatenate(([np.linspace(0, len(covid_de)-1, len(covid_de))], [covid_de.values]), axis=0).T

```

### Perform gradient descent to fit the polynomial function.
### After taking the natural log of the population, the function becomes polynomial.
### Then the polynomial will be transformed back by applying the exponenital function. 

```python
def gradient(y_arr, init_coeff, i):
    grads=[] #All terms in the gradient vector
    '''Add up every term in the polynomial'''
    fixed_term=0
    for j in range(len(init_coeff)):
        fixed_term += init_coeff[j]*y_arr[i, 0]**j
    '''A fixed term that occurs in every polynomial'''
    fixed_term=fixed_term- y_arr[i, 1]
    '''Append all terms in the gradient together'''
    for k in range(len(init_coeff)):
        grads.append(2*fixed_term*y_arr[i, 0]**k)

    grad=np.asarray(grads)
    '''Normalize the gradient vector because only the direction matters.'''
    grad=grad/la.norm(grad)
    return grad

'''Main function to train the coefficients of polynomials'''
def neuro_training(init_coeff, y_arr, error, learn_rate):
    counter=0
    residue_ls = []
    counter_ls = []
    diff = 1
    residue = 0
    '''Iterate until the difference reaches the designated error.'''
    while diff > error:
        residue = 0
        before = init_coeff
        '''Update the gradient for evey data point'''
        for i in range(len(y_arr)):
            grad = gradient(y_arr, init_coeff, i)
            '''Learning rate is 0.01; The step size becomes smaller when iterating'''

            init_coeff = init_coeff - learn_rate*grad/math.log10(counter+2)

            tmpx = 0
            '''Add up the residual error for every data point and every term of polynomial'''
            for j in range(len(init_coeff)):
                 tmpx+= init_coeff[j] * y_arr[i, 0] ** j

            residue += (tmpx - y_arr[i, 1]) ** 2
        after = init_coeff

        diff = la.norm(after - before)
        '''Take the mean square residual error, which is not effected by the sample size'''
        residue = math.sqrt(residue/len(y_arr))
        #print(residue)
        residue_ls.append(residue)
        counter += 1
        counter_ls.append(counter)

    return residue, init_coeff, residue_ls, counter_ls



### Plot the residual error v.s. iteration: not used here.
def re_vs_iter():
    ''''''


    '''Initialization'''
    init_coeff = np.asarray([2., -2., 1.])

    '''Data points with some noises'''
    y_arr = target_data(target_coeff=[-4.5, 1, 3.5], num_data=200, abs_range=1) #+ 0.2*np.random.randn(200, 2)
    residue, init_coeff, residue_ls, counter_ls= neuro_training(init_coeff, y_arr, 2000)

    print("Fitting coefficients:")
    print(init_coeff)
    print()
    print("Root mean square residual error:")
    print(residue)

    '''Plot functions'''
    plt.figure(figsize=(10, 5))
    plt.plot(counter_ls[100:], residue_ls[100:], 'b-')

    plt.xlabel("No. of Iteration")
    plt.ylabel("Residual Error for learning rate 0.01")
    plt.title("Residual Error v.s. Iterations with learning rate 0.01")
    plt.savefig("RE_VS_Iter_arbitrary")
```

### The process of piecewise fitting. 
### Fit the data for a few days of time series at a time, and proceed the fitting forward.

## Piecewise fitting on confirmed cases in Italy 

```python
plt.figure(figsize=(6, 6))
ax = plt.subplot()
ax.plot(covid_it[:,0], np.exp(covid_it[:,1]), "r.", label="Italia")
init_coeff = np.asarray([-0.1, 0.1])
overlap = 1 # The days of prediction that overlap the previous prediction
slen = 3 # The length of the prediction for every segment of time series.

#slen > overlap; overlap > 0 

for i in range(math.floor((len(covid_it)-slen)/overlap)+1):
    if i == math.floor((len(covid_it)-slen)/overlap): #(For the last few points)
        residue, init_coeff,residue_ls, counter_ls=neuro_training(init_coeff, covid_it[overlap*i:, :], 0.005, 0.005)
        ys = [np.dot(init_coeff, [x**i for i in range(len(init_coeff))]) for x in covid_it[overlap*i:, 0]]
        ax.plot(covid_it[overlap*i:, 0], np.exp(ys), "b-")
        ax.annotate(str(i+1), (covid_it[overlap*i, 0], np.exp(ys[0])))
        
    residue, init_coeff,residue_ls, counter_ls=neuro_training(init_coeff, covid_it[overlap*i:(slen+overlap*i), :], 0.005, 0.005)
    ys = [np.dot(init_coeff, [x**i for i in range(len(init_coeff))]) for x in covid_it[overlap*i:(slen+overlap*i), 0]]
    ax.plot(covid_it[overlap*i:(slen+overlap*i), 0], np.exp(ys), "b-")
    ax.annotate(str(i+1), (covid_it[overlap*i, 0], np.exp(ys[0])))
plt.xlabel("Day No.")
plt.ylabel("Numeri di Persone in Italia")
plt.savefig("Italia_Anfitten")
```
