```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
from scipy import stats    
import math
from sklearn.decomposition import PCA
from scipy.interpolate import Rbf
import scipy.cluster.hierarchy as shc
from mpl_toolkits.mplot3d import axes3d, Axes3D
from numpy import linalg as la
```

```python
covid = pd.read_csv("time_series_covid_19_confirmed_de.csv").drop(["Lat", "Long"], axis=1).groupby("Country/Region").sum()
covid_de = covid.loc["Germany",:][1:]
bestaetigt_de_tag=(covid - covid.shift(periods=1, axis=1)).loc["Germany",:][1:]
```

```python
covid = pd.read_csv("time_series_covid_19_deaths_de.csv").drop(["Lat", "Long"], axis=1).groupby("Country/Region").sum()
todesfaelle_de_tag=(covid - covid.shift(periods=1, axis=1)).loc["Germany",:][1:]
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot()
ax.plot(bestaetigt_de_tag.values, "g-")
ax.set_xticks(np.linspace(0, len(covid.columns)-2, len(covid.columns)-1))
ax.set_xticklabels(covid.columns[1:], rotation=90)
ax.set_xlabel("Datum")
ax.set_ylabel("Bestätigte Fälle pro Tag")
ax.set_xlim(38, 77)
plt.savefig("verlangsamt")
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot()
ax.plot(todesfaelle_de_tag.values, "g-")
ax.set_xticks(np.linspace(0, len(covid.columns)-2, len(covid.columns)-1))
ax.set_xticklabels(covid.columns[1:], rotation=90)
ax.set_xlabel("Datum")
ax.set_ylabel("Todesfälle pro Tag")
ax.set_xlim(38, 77)
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot()
ax.plot(np.log10(covid_de.values+1), np.log10(bestaetigt_de_tag.values+1),"g-")

ax.set_xlabel("Gesamte Fälle")
ax.set_ylabel("Bestätigte Fälle pro Tag")
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot()
ax.plot(np.linspace(1, len(covid["4/6/20"]), len(covid["4/6/20"])),np.log10(covid.sort_values("4/6/20",ascending=False)["4/6/20"]),"g.")

ax.set_xlabel("Rang am 06.04.2020")
ax.set_ylabel("Gesamte Todesfälle")
```

```python
plt.figure(figsize=(10, 10))
ax = plt.subplot()

plt.hist(np.log2(covid["4/6/20"]+1), density=True, bins=10)

ax.set_ylabel("Häufigkeit")
ax.set_xlabel("Gesamte Todesfälle")
```
